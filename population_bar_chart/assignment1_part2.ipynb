```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
population = pd.read_csv('p_data.csv', index_col = 'Country Name')
population.drop(['Indicator Name', 'Indicator Code', 'Unnamed: 65'], axis = 1, inplace = True)
population
```

```python
top_countries = population.loc[['China', 'India', 'United States', 'Indonesia', 'Pakistan'], :]
top_countries
```

```python
plt.style.use('seaborn-colorblind')
```

```python
c_code_list = list(top_countries['Country Code'])
c_code_list
```

```python
plt.figure(figsize=(10,12.5))
colors = ['blue', 'red', 'darkslategrey','indigo', 'grey']
plt.plot(top_countries.drop(['Country Code'], axis = 1).T)

decades = ['1960', '1970','1980','1990','2000','2010','2020']
plt.xticks(decades, fontsize = 12)
plt.yticks(fontsize = 12)

plt.legend(c_code_list, fontsize = 12)
plt.title('Top 5 most populated countries', fontsize = 14, weight = 'bold')
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Population in million', fontsize = 14)
plt.grid(True)
plt.show()
```

```python
turkey = population.loc[['Turkey'], :]
f_countries = population.loc[population['2010'] > population.loc['Turkey', '2010']]
f_countries = f_countries.sample(n = 4)
random_countries = pd.concat([turkey, f_countries]) 
random_countries 
```

```python
fr_countries = random_countries.drop(['Country Code'], axis = 1).T
colors = ['slategrey', 'crimson', 'gold','springgreen', 'royalblue']
fr_countries.plot.area(figsize=(10,7.5), title = 'Population of 5 random countries', color = colors)
plt.show()
```
